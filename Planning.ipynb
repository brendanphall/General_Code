```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Define project timelines
data = {
    "Project": [
        "ATFS Migration to GeoServer",
        "Px Migration to Esri",
        "Rj Migration to Esri",
        "BetterRoads App",
        "Fieldwork"
    ],
    "Start Date": [
        "2025-01-29",
        "2025-02-05",
        "2025-02-05",
        "2025-03-01",
        "2025-02-01"
    ],
    "End Date": [
        "2025-05-31",
        "2025-05-31",
        "2025-05-31",
        "2025-06-15",
        "2025-04-30"
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert dates to datetime
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["End Date"] = pd.to_datetime(df["End Date"])

# Plotting the Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each project
for i, row in df.iterrows():
    ax.barh(row["Project"], (row["End Date"] - row["Start Date"]).days, left=row["Start Date"], align='center')

# Format the x-axis as dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.xaxis.set_major_locator(mdates.MonthLocator())

# Add grid, labels, and title
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlabel("Timeline")
plt.ylabel("Projects")
plt.title("Project Timelines for Strategic Planning")
plt.tight_layout()

# Show the chart
plt.show()
```
